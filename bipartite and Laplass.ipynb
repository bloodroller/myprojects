Необходимо было показать, что граф двудольный тогда и только тогда, когда спектр Лапласса равен "беззнаковому" спектру Лапласса. Я брал двудольные графы из сайта http://konect.uni-koblenz.de/networks/.

```python
x = pd.read_csv('out.opsahl-ucforum',sep = ' ',index_col=None)
x= x.iloc[1:,:2].reset_index()
x.columns = ['node1','node2','w']
x.loc[:,'node2'] = x.node2.values + x.node1.unique().shape[0]# Прибавил число вершин в "первой доле",
x.loc[:,'node1'] = x.node1.astype(int)                          #чтобы граф был двудольным.
x.to_csv('n',header=None,sep=' ',index=None)
G = nx.read_edgelist('n',nodetype=int, data=(('weight',float),))
```

Найдем матрицу сопряженности и диагональную матрицу степеней

```python
sh1 = x.shape[0]
sh2 = x.node2.values.max()
A = np.zeros((sh2,sh2))
D = np.zeros((sh2,sh2))
for i in range(sh1):
    A[x.node1.values[i]-1,x.node2.values[i]-1]=1
for k in x.node1.unique():
    D[k-1,k-1] = x[x.node1==k].shape[0]
for k in x.node2.unique():
    D[k-1,k-1] = x[x.node2==k].shape[0]
```

```python
L1 = D - A
L2 = D + A
```

```python
np.sum(L1.diagonal()==L2.diagonal())/L1.diagonal().shape[0]
```
